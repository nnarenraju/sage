```python
import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")

import matplotlib.pyplot as pp
from pycbc.waveform import get_td_waveform
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15.0, 3.0))
for apx in ['IMRPhenomPv2', 'IMRPhenomD', 'IMRPhenomXPHM']:
    if apx == 'IMRPhenomD':
        hp, hc = get_td_waveform(approximant=apx,
                                mass1=25,
                                mass2=24,
                                spin1z=0.05,
                                spin2z=0.8,
                                delta_t=1.0/4096,
                                f_lower=20)
    else:
        hp, hc = get_td_waveform(approximant=apx,
                                mass1=25,
                                mass2=24,
                                spin1x=0.15,
                                spin1y=0.3,
                                spin1z=0.05,
                                spin2x=0.05,
                                spin2y=0.1,
                                spin2z=0.8,
                                delta_t=1.0/4096,
                                f_lower=20)

    hp = hp.time_slice(-2.5, -0.5)
    pp.plot(hp.sample_times, hp, label=apx)

pp.ylabel('Strain')
pp.xlabel('Time (s)')
pp.legend()
pp.show()
```
